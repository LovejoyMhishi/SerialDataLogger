```python
import serial
import csv
import time

# --- Serial setup ---
ser = serial.Serial('COM21', 115200, timeout=5)

# --- CSV setup ---
fieldnames = ["Timestamp", "FenceBreachCount", "Temp", "SecondCount"]

with open("Samples_data_logger.csv", mode="w", newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    timecount = 0 

    while True:
        line = ser.readline().decode('utf-8').strip()
        if line:
            # Split the incoming message by commas
            data = line.split(',')

            # Prepare row
            row = {
                "Timestamp": time.strftime("%Y-%m-%d %H:%M:%S"),
                "FenceBreachCount": data[0],
                "Temp": data[3],
                "SecondCount":timecount    
            }
            print(row)
            # Write row to CSV
            writer.writerow(row)
            csvfile.flush()  # Ensure immediate write to disk

            

```
